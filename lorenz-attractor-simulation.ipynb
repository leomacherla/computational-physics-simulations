```python
import numpy as np
from scipy.integrate import RK45
import matplotlib.pyplot as plt
from collections import deque
import matplotlib.animation as animation
```

# Derivatives

$$ 
\begin{align*}

&\frac{dx}{dt} = \sigma(y - x) \\\\

&\frac{dy}{dt} = x(\rho - z) - y \\\\

&\frac{dz}{dt} = xy - \beta z 

\end{align*}
$$

```python
def derivatives(t, state, sigma, rho, beta):
    x, y, z = state

    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = (x * y) - (beta * z)

    return np.array([dx_dt, dy_dt, dz_dt])
```

# Lorenz Attractor Physics

```python
dt = 0.001 # step size

sigma = 10
rho = 28
beta = 8/3
```

```python
solver = RK45(
    fun=lambda t, state: derivatives(t, state, sigma, rho, beta),
    t0=0,
    y0=[1, 1, 1],
    t_bound=np.inf,
    max_step=dt
)
```

# Simulation

```python
%matplotlib widget

mfig, ax = plt.subplots()

# hide matplotlib widget ui
mfig.canvas.toolbar_visible = False
mfig.canvas.header_visible = False
mfig.canvas.footer_visible = False

# black background colour
ax.set_facecolor('black')
mfig.patch.set_facecolor('black')

# hide axes
ax.set_xticks([])
ax.set_yticks([])
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

# set x and y axes to fit attractor
ax.set_xlim(-20, 20)
ax.set_ylim(0, 50)
ax.set_aspect("equal")

TRACE_LENGTH = 100000
trace_x = deque(maxlen=TRACE_LENGTH)
trace_y = deque(maxlen=TRACE_LENGTH)
trace, = ax.plot([], [], linewidth=0.5, color="#ff0059")

def update(frame): 
    solver.step()
    state = solver.y

    trace_x.append(state[0])
    trace_y.append(state[2])

    trace.set_data(list(trace_x), list(trace_y))

    return trace

anim = animation.FuncAnimation(mfig, update, interval=dt*1000, cache_frame_data=False)

plt.show()
```
