```python
import numpy as np
import plotly.express as px
```

# Constants

```python
g = 9.81 # acceleration due to gravity on Earth (ms^-1)
```

# Euler's Method

$$ x_{n+1} = x_{n} + \Delta t \bigg(\frac{dx}{dt}\bigg)_n $$

```python
def eulers_method(initial_value, step_size, rate_of_change):
    return initial_value + (step_size * rate_of_change)
```

```python
def dv_dt(velocity_x, velocity_y, drag_coefficient, mass):
    speed = np.sqrt(velocity_x**2 + velocity_y**2) # speed of projectile using Pythagoras' theorem

    return ((-drag_coefficient * speed * velocity_x) / mass, ((mass * -g) -drag_coefficient * speed * velocity_y) / mass) # rate of change of x and y velocities
```

# Projectile Physics

```python
dt = 0.01 # step size of 0.01s
t_end = 5 # simulation runs up to 5s
k = 10 # drag coefficient of 10
m = 2 # mass of 2kg

t = np.arange(0, t_end + dt, dt) # each time step
x = np.zeros(len(t)) # x positions
y = np.zeros(len(t)) # y positions
vx = np.zeros(len(t)) # x velocities
vy = np.zeros(len(t)) # y velocities

# initial conditions
vx[0] = 2
vy[0] = 5

for i in np.arange(0, len(t) - 1, 1):

    newY = eulers_method(y[i], dt, vy[i])

    # if y position is on/below ground then stop loop and remove remaining zero elements from arrays
    if(newY < 0):
        x = x[:i+1]
        vx = vx[:i+1]
        y = y[:i+1]
        vy = vy[:i+1]
        t = t[:i+1]
        break

    # use euler's method to compute new x and y positions
    x[i + 1] = eulers_method(x[i], dt, vx[i]) 
    y[i + 1] = newY 

    (dvx_dt, dvy_dt) = dv_dt(vx[i], vy[i], k, m)

    # use euler's method to compute new x and y velocities
    vx[i + 1] = eulers_method(vx[i], dt, dvx_dt) 
    vy[i + 1] = eulers_method(vy[i], dt, dvy_dt)
```

```python
# plot scatter graph of x and y values
px.scatter(x=x, y=y, template="plotly_dark", title=f"x and y values for a projectile, mass {m}kg, launched with velocity ({vx[0]}, {vy[0]})ms<sup>-1</sup>, with a drag coefficient of {k}")
```
